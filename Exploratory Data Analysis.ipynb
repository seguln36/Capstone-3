```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('returns.csv')
df.head(50)
```

```python
df.drop(['Unnamed: 0'], axis=1)
```

```python
df.info()
```

```python
plt.figure(figsize=(25, 10));
plt.plot(df.Open_AMZN.values, color='red', label='open AMZN')
plt.plot(df.Close_AMZN.values, color='green', label='close AMZN')
plt.plot(df.Open_AAPL.values, color='blue', label='open AAPL')
plt.plot(df.Close_AAPL.values, color='purple', label='close AAPL')
plt.plot(df.Open_AMD.values, color='black', label='open AMD')
plt.plot(df.Close_AMD.values, color='yellow', label='close AMD')
plt.plot(df.Open_MSFT.values, color='blue', label='open MSFT')
plt.plot(df.Close_MSFT.values, color='orange', label='close MSFT')
plt.plot(df.Open_NVDA.values, color='red', label='open NVDA')
plt.plot(df.Close_NVDA.values, color='black', label='close NVDA')
#plt.plot(df.Return_AMZN.values, color='blue', label='return')

#plt.plot(df_stock_norm.volume.values, color='gray', label='volume')
plt.title('stock')
plt.xlabel('time [days]')
plt.ylabel('normalized price/volume')
plt.legend(loc='best')
plt.show()
```

```python
sns.set()
for i in [df['Return_AMZN'],df['Return_AAPL'],df['Return_MSFT'],df['Return_NVDA'],df['Return_AMD']]:
    plt.figure(figsize=(40,20))
    sns.scatterplot(df['Date'],i)
    plt.show()
```

```python
plt.figure(figsize=(40,20))
for i in [df['Close_AMZN'],df['Close_AAPL'],df['Close_MSFT'],df['Close_NVDA'],df['Close_AMD']]:
    sns.scatterplot(df['Date'],i)
    plt.show()
```

```python
sns.boxplot(x=df['Date'],y=df['Return_AMZN'])
sns.boxplot(x=df['Date'],y=df['Return_AAPL'])
sns.boxplot(x=df['Date'],y=df['Return_MSFT'])
sns.boxplot(x=df['Date'],y=df['Return_NVDA'])
sns.boxplot(x=df['Date'],y=df['Return_AMD'])
```

```python
plt.figure(figsize=(25,25))
corrMatrix = df.corr()
sns.heatmap(corrMatrix, annot=True, cmap='RdBu')

plt.show()
```

```python
sns.distplot(df.loc['2010-05-01':'2020-04-30']['Return_AMZN'],color='green',bins=100)
```

```python
#highlighting seasonality
plt.figure(figsize=(12,6))
#df = df.loc['2010-01-04':'2012-01-04']
mask = (df['Date'] > '2012-01-01') & (df['Date'] <= '2013-01-01')

df['Return_AMZN'].loc[mask].rolling(window=30).mean().plot(label='30 Day Avg')
df['Return_AMZN'].loc[mask].plot(label='AMZN CLOSE')
plt.legend()

mask = (df['Date'] > '2015-01-01') & (df['Date'] <= '2016-01-01')

plt.figure(figsize=(12,6))
df['Return_AMZN'].loc[mask].rolling(window=30).mean().plot(label='30 Day Avg')
df['Return_AMZN'].loc[mask].plot(label='AMZN CLOSE')
plt.legend()

mask = (df['Date'] > '2019-01-01') & (df['Date'] <= '2020-01-01')                    
plt.figure(figsize=(12,6))
df['Return_AMZN'].loc[mask].rolling(window=30).mean().plot(label='30 Day Avg')
df['Return_AMZN'].loc[mask].plot(label='AMZN CLOSE')
plt.legend()
```

```python
df['Return_AMZN'].loc[mask].ta_plot(study='sma',periods=[13,21,55],title='Simple Moving Averages')
```
