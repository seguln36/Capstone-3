```python
import yfinance
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ticker = yfinance.Ticker('AMZN')
AMZN = ticker.history(interval="1d",
    start="2019-01-01",end="2019-12-31")
AMZN
```

```python
stocks = ['AMZN','AAPL','MSFT','NVDA','AMD']
returns = pd.DataFrame({})
for t in stocks:
    #name = t
    ticker = yfinance.Ticker(t)
    data = ticker.history(interval="1d",
    start="2010-01-01",end="2019-12-31")
    data['Close_%s' % t] = data['Close']
    data['Open_%s' % t] = data['Open']
    data['Volume_%s' % t] = data['Volume']
    data['Return_%s' % t] = data['Close'].pct_change(1)
    
    returns = returns.append(data['Return_%s' % t])
    returns = returns.append(data['Volume_%s' % t])
    returns = returns.append(data['Close_%s' % t])
    returns = returns.append(data['Open_%s' % t])
    
```

```python
returns.head()
```

```python
returns = returns.T
```

```python
returns = returns.reset_index()
returns = returns.rename(columns={'index': 'Date'})
returns
```

```python
returns.to_csv('returns.csv')
```
